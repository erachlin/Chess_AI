```python
def get_coord(square):
    file = ord(square[0]) - 97
    if square[1] == '1':
        rank = 7
    elif square[1] == '2':
        rank = 6
    elif square[1] == '3':
        rank = 5
    elif square[1] == '4':
        rank = 4
    elif square[1] == '5':
        rank = 3
    elif square[1] == '6':
        rank = 2
    elif square[1] == '7':
        rank = 1
    else:
        rank = 0
    #print(square, rank, file)
    return rank, file

def get_square(x, y):
    file = chr(y + 97)
    if x == 0:
        rank = '8'
    elif x == 1:
        rank = '7'
    elif x == 2:
        rank = '6'
    elif x == 3:
        rank = '5'
    elif x == 4:
        rank = '4'
    elif x == 5:
        rank = '3'
    elif x == 6:
        rank = '2'
    else:
        rank = '1'
    return file + rank
```

```python
def make_legal_moves(input_state, action_space, depth):
    depth += 1
    turn = input_state.current_player
    check = False
#     if turn == 'p1':
#         opp = 'p2'
#     else:
#         opp = 'p1'
    current_king = ""
    p1_legal_moves = []
    p2_legal_moves = []
    occ = input_state.square_occupation
    for square in action_space:
        file = square[0]
        rank = square[1]
        piece = ''
        x,y = get_coord(square)
        pieces = input_state.get_pieces()
        for item in pieces:
            plane = pieces[item]

            if plane[x][y] == 1:
                piece = item[-1]
                
                #todo: en-passant
                if piece == 'P':
                    if item[:2] == 'p1':
                        if occ[x - 1][y] == 0 and square[1] != '7':
                            
                            p1_legal_moves.append((file + str(int(rank) + 1), action_space[square]['N1'], square))
                            
                            if square[1] == '2' and occ[x - 2][y] == 0:
                            
                                p1_legal_moves.append((file + '4', action_space[square]['N2'], square))
                                
                                             
                        if y >= 1 and occ[x - 1][y - 1] == 2 and square[1] != '7':
                            
                            p1_legal_moves.append((file + 'x' + chr(ord(file) - 1) + str(int(rank) + 1), action_space[square]['NW1'], square))
                        
                        if y <= 6 and occ[x - 1][y + 1] == 2 and square[1] != '7':
                            
                            p1_legal_moves.append((file + 'x' + chr(ord(file) + 1) + str(int(rank) + 1), action_space[square]['NE1'], square))
                        
                        elif square[1] == '7':
                                               
                            if occ[0][y] == 0:

                                p1_legal_moves.append((file + '8=Q', action_space[square]['N1'], square))

                                p1_legal_moves.append((file + '8=N', action_space[square]['KPF'], square))
                                p1_legal_moves.append((file + '8=B', action_space[square]['BPF'], square))
                                p1_legal_moves.append((file + '8=R', action_space[square]['RPF'], square))
                                
                            
                            if y >= 1 and occ[0][y - 1] == 2:
                                
                                p1_legal_moves.append((file + 'x' + chr(ord(file) - 1) + '8=Q', action_space[square]['NW1'], square))

                                p1_legal_moves.append((file + 'x' + chr(ord(file) - 1) + '8=N', action_space[square]['KPW'], square))
                                p1_legal_moves.append((file + 'x' + chr(ord(file) - 1) + '8=B', action_space[square]['BPW'], square))
                                p1_legal_moves.append((file + 'x' + chr(ord(file) - 1) + '8=R', action_space[square]['RPW'], square))
                            
                            if y <= 6 and occ[0][y + 1] == 2:
                                p1_legal_moves.append((file + 'x' + chr(ord(file) + 1) + '8=Q', action_space[square]['NE1'], square))

                                p1_legal_moves.append((file + 'x' + chr(ord(file) + 1) + '8=N', action_space[square]['KPE'], square))
                                p1_legal_moves.append((file + 'x' + chr(ord(file) + 1) + '8=B', action_space[square]['BPE'], square))
                                p1_legal_moves.append((file + 'x' + chr(ord(file) + 1) + '8=R', action_space[square]['RPE'], square))
                        
                        
                                               
                    else:
                        if occ[x + 1][y] == 0 and square[1] != '2':
                            
                            p2_legal_moves.append((file + str(int(rank) - 1), action_space[square]['S1'], square))
    
                            if square[1] == '7' and occ[x + 2][y] == 0:
                            
                                p2_legal_moves.append((file + '5', action_space[square]['S2'], square))


                        if y >= 1 and occ[x + 1][y - 1] == 1 and square[1] != '2':
                            
                            p2_legal_moves.append((file + 'x' + chr(ord(file) - 1) + str(int(rank) - 1), action_space[square]['SW1'], square))
                        
                        if y <= 6 and occ[x + 1][y + 1] == 1 and square[1] != '2':
                            
                            p2_legal_moves.append((file + 'x' + chr(ord(file) + 1) + str(int(rank) - 1), action_space[square]['SE1'], square))
                        
                        elif square[1] == '2':
                                               
                            if occ[7][y] == 0:

                                p2_legal_moves.append((file + '1=Q', action_space[square]['S1'], square))

                                p2_legal_moves.append((file + '1=N', action_space[square]['KPF'], square))
                                p2_legal_moves.append((file + '1=B', action_space[square]['BPF'], square))
                                p2_legal_moves.append((file + '1=R', action_space[square]['RPF'], square))
                            
                            if y >= 1 and occ[7][y - 1] == 1:
                                
                                p2_legal_moves.append((file + 'x' + chr(ord(file) - 1) + '1=Q', action_space[square]['SW1'], square))

                                p2_legal_moves.append((file + 'x' + chr(ord(file) - 1) + '1=N', action_space[square]['KPW'], square))
                                p2_legal_moves.append((file + 'x' + chr(ord(file) - 1) + '1=B', action_space[square]['BPW'], square))
                                p2_legal_moves.append((file + 'x' + chr(ord(file) - 1) + '1=R', action_space[square]['RPW'], square))
                            
                            if y <= 6 and occ[7][y + 1] == 1:
                                p2_legal_moves.append((file + 'x' + chr(ord(file) + 1) + '1=Q', action_space[square]['SE1'], square))

                                p2_legal_moves.append((file + 'x' + chr(ord(file) + 1) + '1=N', action_space[square]['KPE'], square))
                                p2_legal_moves.append((file + 'x' + chr(ord(file) + 1) + '1=B', action_space[square]['BPE'], square))
                                p2_legal_moves.append((file + 'x' + chr(ord(file) + 1) + '1=R', action_space[square]['RPE'], square))
                        
                        
                
                elif piece == 'N':
                    if item[:2] == 'p1':
                        knight = np.asarray(list(action_space[square].values())[-8:])
                        for move in knight:
                            try:
                                rank = np.nonzero(move)[0][0]
                                file = np.nonzero(move)[1][0]
                                new_square = get_square(rank, file)
                                if occ[rank][file] == 0:
                                    p1_legal_moves.append(('N' + new_square, move, square))
                                elif occ[rank][file] == 2:
                                    p1_legal_moves.append(('Nx' + new_square, move, square))
                            except:
                                continue
                    else:
                        knight = np.asarray(list(action_space[square].values())[-8:])
                        for move in knight:
                            try:
                                                          
                                rank = np.nonzero(move)[0][0]
                                file = np.nonzero(move)[1][0]
                                new_square = get_square(rank, file)
                                if occ[rank][file] == 0:
                                    p2_legal_moves.append(('N' + new_square, move, square))
                                elif occ[rank][file] == 1:
                                    p2_legal_moves.append(('Nx' + new_square, move, square))
                            
                            except:
                                continue
                
                elif piece == 'B':
                    bishop = {'NE' : [], 'SE' : [], 'SW' : [], 'NW' : []}
                    for key in action_space[square]:
                        if any(char.isdigit() for char in key) and len(key) == 3:
                            bishop[key[:2]].append(np.asarray(action_space[square][key]))
                        
                    if item[:2] == 'p1':
                        
                        for key in bishop:
                            for move in bishop[key]:
                                try:
                                    rank = np.nonzero(move)[0][0]
                                    file = np.nonzero(move)[1][0]
                                    new_square = get_square(rank, file)
                                    if occ[rank][file] == 0:
                                        p1_legal_moves.append((piece + new_square, move, square))
                                    elif occ[rank][file] == 2:
                                        p1_legal_moves.append((piece + 'x' + new_square, move, square))
                                        break
                                    else:
                                        break
                                except:
                                    break
                    else:
                        for key in bishop:
                            for move in bishop[key]:
                                try:
                                    rank = np.nonzero(move)[0][0]
                                    file = np.nonzero(move)[1][0]
                                    new_square = get_square(rank, file)
                                    if occ[rank][file] == 0:
                                        p2_legal_moves.append((piece + new_square, move, square))
                                    elif occ[rank][file] == 1:
                                        p2_legal_moves.append((piece +  'x' + new_square, move, square))
                                        break
                                    else:
                                        break
                                except:
                                    break
                
                elif piece == 'R':
                    rook = {'N' : [], 'S' : [], 'W' : [], 'E' : []}
                    for key in list(action_space[square].keys())[:28]:
                        rook[key[0]].append(np.asarray(action_space[square][key]))
                    
                    if item[:2] == 'p1':
                    
                        for key in rook:
                            for move in rook[key]:
                                try:
                                    rank = np.nonzero(move)[0][0]
                                    file = np.nonzero(move)[1][0]
                                    new_square = get_square(rank, file)
                                    if occ[rank][file] == 0:
                                        p1_legal_moves.append((piece + new_square, move, square))
                                    elif occ[rank][file] == 2:
                                        p1_legal_moves.append((piece + 'x' + new_square, move, square))
                                        break
                                    else:
                                        break
                                except:
                                    break
                        
                    else:
                        for key in rook:
                            for move in rook[key]:
                                try:
                                    rank = np.nonzero(move)[0][0]
                                    file = np.nonzero(move)[1][0]
                                    new_square = get_square(rank, file)
                                    if occ[rank][file] == 0:
                                        p2_legal_moves.append((piece + new_square, move, square))
                                    elif occ[rank][file] == 1:
                                        p2_legal_moves.append((piece + 'x' + new_square, move, square))
                                        break
                                    else:
                                        break
                                except:
                                    break
                                                          
                elif piece == 'Q':
                    bishop_moves = {'NE' : [], 'SE' : [], 'SW' : [], 'NW' : []}
                    rook_moves = {'N' : [], 'S' : [], 'W' : [], 'E' : []}
                    
                    
                    for key in list(action_space[square].keys())[:28]:
                        rook_moves[key[0]].append(np.asarray(action_space[square][key]))
                    
                    for key in action_space[square]:
                        if any(char.isdigit() for char in key) and len(key) == 3:
                            bishop_moves[key[:2]].append(np.asarray(action_space[square][key]))
                        
                    if item[:2] == 'p1':
                        
                        for key in bishop_moves:
                            for move in bishop_moves[key]:
                                try:
                                    rank = np.nonzero(move)[0][0]
                                    file = np.nonzero(move)[1][0]
                                    new_square = get_square(rank, file)
                                    if occ[rank][file] == 0:
                                        p1_legal_moves.append((piece + new_square, move, square))
                                    elif occ[rank][file] == 2:
                                        p1_legal_moves.append((piece + 'x' + new_square, move, square))
                                        break
                                    else:
                                        break
                                except:
                                    break
                        
                                            
                        for key in rook_moves:
                            for move in rook_moves[key]:
                                try:
                                    rank = int(np.nonzero(move)[0][0])
                                    file = int(np.nonzero(move)[1][0])
                                    new_square = get_square(rank, file)
                                    if occ[rank][file] == 0:
                                        p1_legal_moves.append((piece + new_square, move, square))
                                    elif occ[rank][file] == 2:
                                        p1_legal_moves.append((piece + 'x' + new_square, move, square))
                                        break
                                    else:
                                        break
                                except:
                                    break
                    else:
                        for key in bishop_moves:
                            for move in bishop_moves[key]:
                                try:
                                    rank = np.nonzero(move)[0][0]
                                    file = np.nonzero(move)[1][0]
                                    new_square = get_square(rank, file)
                                    if occ[rank][file] == 0:
                                        p2_legal_moves.append((piece + new_square, move, square))
                                    elif occ[rank][file] == 1:
                                        p2_legal_moves.append((piece +  'x' + new_square, move, square))
                                        break
                                    else:
                                        break
                                except:
                                    break
            
                        
                        for key in rook_moves:
                            for move in rook_moves[key]:
                                try:
                                    rank = np.nonzero(move)[0][0]
                                    file = np.nonzero(move)[1][0]
                                    new_square = get_square(rank, file)
                                    if occ[rank][file] == 0:
                                        p2_legal_moves.append((piece + new_square, move, square))
                                    elif occ[rank][file] == 1:
                                        p2_legal_moves.append((piece + 'x' + new_square, move, square))
                                        break
                                    else:
                                        break
                                except:
                                    break
                    
                        
                elif piece == 'K':
                    king = []
                    if turn == item[:2]:
                        current_king = square
                        
                    for key in action_space[square]:
                        if len(key) > 1 and key[-2] != '1' and key[-1] == 1:
                            king.append(np.asarray(action_space[square][key], square))
                        
                    if item[:2] == 'p1':
                        
                        for move in king:
                            try:
                                rank = np.nonzero(move)[0][0]
                                file = np.nonzero(move)[1][0]
                                new_square = get_square(rank, file)
                                if occ[rank][file] == 0:
                                    p1_legal_moves.append((piece + new_square, move, square))
                                elif occ[rank][file] == 2:
                                    p1_legal_moves.append((piece + 'x' + new_square, move, square))
                            
                            except:
                                continue
                    
                    else:
                        
                        for move in king:
                            try:
                                rank = np.nonzero(move)[0][0]
                                file = np.nonzero(move)[1][0]
                                new_square = get_square(rank, file)
                                if occ[rank][file] == 0:
                                    p2_legal_moves.append((piece + new_square, move, square))
                                elif occ[rank][file] == 1:
                                    p2_legal_moves.append((piece + 'x' + new_square, move, square))
                            
                            except:
                                continue
                                                     
                                                              
    checking for check
    if turn == 0:
        for move in p2_legal_moves:
            if move[1][-2:] == current_king:
                self.check = True
    else:
        for move in p1_legal_moves:
            if move[1][-2:] == current_king:
                self.check = True
            
    
    legal_moves = []
    if turn == 0:
        for move in p1_legal_moves:
            if not_check(input_state, move, current_king, depth):
                legal_moves.append(move)
    else:
        for move in p2_legal_moves:
            if not_check(input_state, move, current_king, depth):
                legal_moves.append(move)
 
    #print(current_king)
    return legal_moves
                                                              
    
def not_check(game, move, king_location, depth):
    if depth > 2 or len(game.history) <= 2:
        return True
    scratch = game.clone()
    scratch.apply(move)
    scratch.make_image(len(scratch.history))
    scratch.current_player = len(scratch.history) % 2
    opp_legal_moves = make_legal_moves(scratch, possible_moves, depth)
    for move in opp_legal_moves:
        if move[1][-2:] == current_king:
             return False

    return True
```
